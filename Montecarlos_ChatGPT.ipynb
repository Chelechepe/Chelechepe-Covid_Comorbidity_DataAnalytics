```python
import random
import numpy as np
import matplotlib.pyplot as plt

# Define the number of iterations for the simulation
num_iterations = 10000

# Set the parameters for the normal distribution of patient visits
avg_visits = 4
std_visits = 1

# Set the earnings per patient visit
earnings_per_visit = 10.00

# Define the number of working days in each iteration
working_days = 22

# Define the number of doctors
num_doctors = 10

# Initialize a list to store the total earnings for each iteration
total_earnings_by_iteration = []

# Run the simulation for the defined number of iterations
for i in range(num_iterations):
    # Generate a random number of patient visits from a normal distribution
    num_visits = int(round(np.random.normal(avg_visits, std_visits)))
    # Calculate the number of patients attended by each doctor in this iteration
    num_patients_by_doctor = [0] * num_doctors
    for j in range(num_visits):
        doctor = j % num_doctors
        num_patients_by_doctor[doctor] += 1
    # Calculate the earnings for this iteration based on the number of patients attended
    earnings = sum(num_patients_by_doctor) * earnings_per_visit
    # Multiply the earnings by the number of working days
    total_earnings = earnings * working_days
    total_earnings_by_iteration.append(total_earnings)

# Generate a histogram of the total earnings by iteration
plt.hist(total_earnings_by_iteration, bins=50, density=True, alpha=0.5, color='green')
plt.xlabel('Total Earnings over {} Days'.format(working_days))
plt.ylabel('Frequency')
plt.title('Earnings Distribution for {} Iterations'.format(num_iterations))
plt.show()
```
